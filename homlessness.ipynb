```python
import pandas as pd
import numpy as np
import autoviz as viz
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

```python
#!pip install autoviz
```

```python
hic_2021 = pd.read_csv('2021-HIC-Counts-by-State.csv')
```

```python
hic_2021
```

```python
df = hic_2021
```

```python
filename = "2021-HIC-Counts-by-State.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=30_000,
    max_cols_analyzed=100,
    save_plot_dir=None
)
```
